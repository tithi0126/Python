```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('iris.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
df['Species']=le.fit_transform(df['Species'])
df.head()
```

```python
# y=f(x)
x=df[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm',	'PetalWidthCm']]
y=df['Species']
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
trainx, testx, trainy, testy=train_test_split(x,y, test_size=0.2, random_state=42, stratify=y)
testx
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model=KNeighborsClassifier(n_neighbors=9, metric='eucladian', weights='distance')
```

```python
model.fit(trainx,trainy)
```

```python
ypred=model.predict(testx)
ypred
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
a=accuracy_score(ypred,testy)*100
a
```

```python
cm=confusion_matrix(ypred,testy)
cm
```
