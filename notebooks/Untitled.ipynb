```python
import pandas as pd
from automl_quickpipe import AutoML

# Load data
df = pd.read_csv('./datasets/heart.csv')

# Create and run AutoML
ml = AutoML(
    df=df,
    target_col='target_variable',
    test_size=0.2,
    problem_type='auto'  # Automatically detects classification/regression
)

# Run full analysis
ml.run_complete_analysis()

# Get best model info
print(f"\nBest Model: {ml.best_model_name}")
print(f"Best Score: {ml.results[ml.best_model_name]}")

# Make predictions on new data if needed
# predictions = ml.predict_new(new_data)
```
